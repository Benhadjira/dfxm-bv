# Predict Burgers Vector for Experimental 111 Reflection

Run `crystallographic_resnet_models.ipynb` first so `benchmark_artifacts/crystallographic_resnet18.pt` exists. This notebook predicts the Burgers vector for the six experimental images using `n = (1, -1, -1)`.

```python
from bv_repro.training import BenchmarkConfig, predict_burgers_vectors
from bv_repro.visualization import plot_experimental_prediction_comparison, plot_experimental_topk_heatmaps

image_paths = [
    "/home/esrf/benhadji/GO-ML-BV/expr_111refl.npy",
    "/home/esrf/benhadji/GO-ML-BV/expr_111refl_2.npy",
    "/home/esrf/benhadji/GO-ML-BV/expr_111refl_3.npy",
    "/home/esrf/benhadji/GO-ML-BV/expr_111refl_4.npy",
    "/home/esrf/benhadji/GO-ML-BV/expr_111refl_5.npy",
    "/home/esrf/benhadji/GO-ML-BV/expr_111refl_6.npy",
]

normal = (1, -1, -1)
config = BenchmarkConfig(
    data_path="/data/projects/engage_id03/int_fcc_002/",
    artifact_dir="benchmark_artifacts",
)
```

```python
crystal_predictions = predict_burgers_vectors(
    image_paths=image_paths,
    normal=normal,
    config=config,
    artifact_dir="benchmark_artifacts",
    model_type="crystallographic",
    topk=3,
)

data_predictions = predict_burgers_vectors(
    image_paths=image_paths,
    normal=normal,
    config=config,
    artifact_dir="benchmark_artifacts",
    model_type="data",
    topk=3,
)

for crystal_pred, data_pred in zip(crystal_predictions, data_predictions):
    print(crystal_pred["image_path"])
    print("  Crystallographic constrained")
    print(f"    predicted b = {crystal_pred['predicted_burgers']}  confidence = {crystal_pred['confidence']:.4f}")
    print("    top candidates:")
    for cand in crystal_pred["topk"]:
        print(f"      b = {cand['burgers']}  p = {cand['probability']:.4f}")

    print("  Data-driven")
    print(f"    predicted b = {data_pred['predicted_burgers']}  confidence = {data_pred['confidence']:.4f}")
    print("    top candidates:")
    for cand in data_pred["topk"]:
        print(f"      b = {cand['burgers']}  p = {cand['probability']:.4f}")
    print()
```

```python
plot_experimental_prediction_comparison(
    crystal_predictions,
    data_predictions,
    output_path="experimental_burgers_prediction_comparison.png",
)

plot_experimental_topk_heatmaps(
    crystal_predictions,
    data_predictions,
    output_path="experimental_burgers_topk_heatmaps.png",
)
```
