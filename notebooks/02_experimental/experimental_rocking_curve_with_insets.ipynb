# Experimental rocking curve with image insets

This notebook loads an experimental EDF rocking-curve stack, computes the ROI intensity versus rotation angle, and plots a publication-style rocking curve with image insets placed along the curve. One selected weak-beam frame is highlighted with a red border and arrow.

```python
from pathlib import Path
import re

import matplotlib.pyplot as plt
import numpy as np
from matplotlib.offsetbox import AnnotationBbox, OffsetImage
from matplotlib.patches import Rectangle

from bv_repro.data import read_edf_image, stretch_vertical
```

```python
# -------------------------------------------------------------------------
# User settings
# -------------------------------------------------------------------------
DATA_DIR = Path("dfxm-go-ml/ex_data")

# The loader below uses the first candidate that exists and contains EDF files.
# It searches recursively, so DATA_DIR can be a parent folder.
DATA_DIR_CANDIDATES = [
    DATA_DIR,
    Path("/data/projects/engage_id03/dfxm-go-ml/ex_data"),
    Path("/mnt/multipath-shares/data/projects/engage_id03/dfxm-go-ml/ex_data"),
]


# Use None for the full image, or set (x0, y0, x1, y1) in stretched-image pixels.
# Example from previous ROI work:
ROI_XYXY = None
# ROI_XYXY = (960, 3586, 992, 3668)

# Optional center crop if ROI_XYXY is None and the full image is too large.
# Set to None to use the full frame. Use (height, width), e.g. (1000, 1000).
CENTER_CROP_SHAPE = (1200, 1200)

# EDF/image preprocessing.
APPLY_VERTICAL_STRETCH = True
STRETCH_ANGLE_DEG = 18.0
SUBTRACT_FRAME_MEAN = False

# Rocking-axis calibration.
# Prefer reading the rocking angle from EDF headers/motor positions and sorting numerically.
READ_PHI_FROM_EDF = True

# Leave both as None for automatic detection of the varying rocking motor.
# After running the header-inspection cell, set one of these if needed, e.g. PHI_MOTOR_NAME = "diffrz".
PHI_HEADER_KEY = None
PHI_MOTOR_NAME = "diffry"

PHI_ZERO_MODE = "first"      # "first", "min", or None
PHI_PLOT_SCALE = 1e4         # if EDF phi is in degrees, this plots units of 10^-4 deg
PHI_XLABEL = r"$\phi$ Rotation [$10^{-4}$ deg]"

# Fallback if phi is missing from EDF headers.
PHI_START_MDEG = -1.0
PHI_STEP_MDEG = 1.0

# Image inset frames. If None, frames are chosen automatically.
INSET_FRAMES = None
# INSET_FRAMES = [0, 1, 3, 4, 5, 7, 9]

# Highlighted weak-beam frame.
HIGHLIGHT_FRAME = 5

# Display settings.
DISPLAY_LOG = True
DISPLAY_PERCENTILES = (1.0, 99.8)
CMAP = "gray"
# Insets use the whole selected frame/ROI, only resized smaller.
# Do not crop the inset image; set a small max size instead.
INSET_MAX_PIXELS = (95, 95)         # scaled 1200x1200 ROI inset size: (height, width)
INSET_ZOOM = 0.55

# Output.
OUTPUT_DIR = Path("experimental_rocking_curve_insets")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
OUTPUT_PNG = OUTPUT_DIR / "experimental_rocking_curve_with_insets.png"
OUTPUT_PDF = OUTPUT_DIR / "experimental_rocking_curve_with_insets.pdf"
```

```python
def center_crop_bounds(image_shape, crop_shape):
    h, w = image_shape[:2]
    crop_h, crop_w = crop_shape
    crop_h = min(int(crop_h), h)
    crop_w = min(int(crop_w), w)
    y0 = (h - crop_h) // 2
    x0 = (w - crop_w) // 2
    return x0, y0, x0 + crop_w, y0 + crop_h


def crop_stack(stack, roi_xyxy=None, center_crop_shape=None):
    if roi_xyxy is None and center_crop_shape is None:
        return stack, (0, 0, stack.shape[2], stack.shape[1])

    if roi_xyxy is None:
        roi_xyxy = center_crop_bounds(stack.shape[1:], center_crop_shape)

    x0, y0, x1, y1 = [int(v) for v in roi_xyxy]
    x0 = max(0, min(stack.shape[2] - 1, x0))
    x1 = max(x0 + 1, min(stack.shape[2], x1))
    y0 = max(0, min(stack.shape[1] - 1, y0))
    y1 = max(y0 + 1, min(stack.shape[1], y1))
    return stack[:, y0:y1, x0:x1], (x0, y0, x1, y1)


def subtract_frame_mean(stack):
    stack = np.asarray(stack, dtype=np.float32)
    corrected = stack - stack.mean(axis=(1, 2), keepdims=True)
    return np.clip(corrected, a_min=0.0, a_max=None)


def prepare_display_image(image, log_scale=True, percentiles=(1.0, 99.8)):
    image = np.asarray(image, dtype=np.float32)
    if log_scale:
        image = np.log1p(np.clip(image, a_min=0.0, a_max=None))
    vmin, vmax = np.percentile(image, percentiles)
    if np.isclose(vmin, vmax):
        vmax = vmin + 1.0
    image = np.clip(image, vmin, vmax)
    return (image - vmin) / (vmax - vmin)



def center_crop_image(image, crop_shape):
    if crop_shape is None:
        return image
    x0, y0, x1, y1 = center_crop_bounds(image.shape, crop_shape)
    return image[y0:y1, x0:x1]


def resize_image_for_display(image, max_shape=(120, 180)):
    image = np.asarray(image, dtype=np.float32)
    max_h, max_w = max_shape
    h, w = image.shape[:2]
    scale = min(max_h / h, max_w / w, 1.0)
    if scale >= 1.0:
        return image
    try:
        from scipy.ndimage import zoom
    except ImportError:
        # Fallback: simple strided downsampling.
        step = int(np.ceil(1.0 / scale))
        return image[::step, ::step]
    return zoom(image, zoom=(scale, scale), order=1).astype(np.float32)


def prepare_inset_image(image):
    # Keep the whole selected frame/ROI, only downsample for display.
    image = prepare_display_image(
        image,
        log_scale=DISPLAY_LOG,
        percentiles=DISPLAY_PERCENTILES,
    )
    return resize_image_for_display(image, INSET_MAX_PIXELS)

def choose_inset_frames(num_frames, highlight_frame):
    if num_frames <= 8:
        frames = list(range(num_frames))
    else:
        frames = np.linspace(0, num_frames - 1, 8, dtype=int).tolist()
    if highlight_frame not in frames:
        frames.append(highlight_frame)
    return sorted(set(int(f) for f in frames if 0 <= int(f) < num_frames))



def natural_sort_key(path):
    parts = re.split(r"(\d+(?:\.\d+)?)", Path(path).name)
    key = []
    for part in parts:
        try:
            key.append(float(part))
        except ValueError:
            key.append(part.lower())
    return key


def filename_numeric_key(path):
    numbers = re.findall(r"[-+]?\d*\.\d+|[-+]?\d+", Path(path).stem)
    return float(numbers[-1]) if numbers else np.inf


def find_edf_files_recursive(data_dir):
    data_dir = Path(data_dir)
    patterns = ("*.edf", "*.EDF", "*.edf.gz", "*.EDF.gz")

    files = []
    for pattern in patterns:
        files.extend(data_dir.glob(pattern))
    if files:
        return sorted(files, key=natural_sort_key)

    for pattern in patterns:
        files.extend(data_dir.rglob(pattern))
    return sorted(files, key=natural_sort_key)


def resolve_data_dir_and_files(candidates):
    checked = []
    for candidate in candidates:
        candidate = Path(candidate)
        files = find_edf_files_recursive(candidate) if candidate.exists() else []
        checked.append((candidate, len(files)))
        if files:
            return candidate, files

    message = "No EDF files found. Checked:\n"
    message += "\n".join(f"  {path}  ({count} EDF files)" for path, count in checked)
    raise FileNotFoundError(message)


def _as_float(value):
    if value is None:
        return None
    if isinstance(value, (int, float, np.integer, np.floating)):
        return float(value)
    text = str(value).strip()
    match = re.search(r"[-+]?\d*\.\d+(?:[eE][-+]?\d+)?|[-+]?\d+(?:[eE][-+]?\d+)?", text)
    return float(match.group(0)) if match else None


def _lower_header(header):
    return {str(key).lower(): value for key, value in header.items()}


def _motor_table_from_header(header):
    lower_header = _lower_header(header)
    motor_names = None
    motor_positions = None

    for names_key in ["motor_mne", "motor_names", "counter_mne"]:
        if names_key in lower_header:
            motor_names = str(lower_header[names_key]).replace(",", " ").split()
            break
    for pos_key in ["motor_pos", "motor_positions", "counter_pos"]:
        if pos_key in lower_header:
            motor_positions = str(lower_header[pos_key]).replace(",", " ").split()
            break

    if not motor_names or not motor_positions:
        return {}

    table = {}
    for index, name in enumerate(motor_names):
        if index < len(motor_positions):
            table[name] = _as_float(motor_positions[index])
    return table


def _header_numeric_candidates(header):
    lower_header = _lower_header(header)
    candidates = {}
    interesting_words = ["phi", "rot", "rock", "omega", "th", "eta", "diffrz", "sam"]
    for key, value in lower_header.items():
        key_text = str(key).lower()
        if any(word in key_text for word in interesting_words):
            number = _as_float(value)
            if number is not None:
                candidates[f"header:{key}"] = number

    for motor_name, value in _motor_table_from_header(header).items():
        if value is not None:
            candidates[f"motor:{motor_name}"] = value

    return candidates


def choose_phi_series(headers):
    """Choose the phi/rocking series from EDF headers.

    Priority:
    1. Explicit PHI_HEADER_KEY
    2. Explicit PHI_MOTOR_NAME
    3. Automatic: varying candidate with phi/rot/rock/diffrz-like name, largest finite range
    4. Any varying motor/header candidate, largest finite range
    """
    if PHI_HEADER_KEY is not None:
        values = []
        for header in headers:
            lower_header = _lower_header(header)
            value = lower_header.get(str(PHI_HEADER_KEY).lower())
            values.append(np.nan if value is None else _as_float(value))
        return np.asarray(values, dtype=float), f"header:{PHI_HEADER_KEY}", []

    if PHI_MOTOR_NAME is not None:
        values = []
        for header in headers:
            table = _motor_table_from_header(header)
            lower_table = {name.lower(): value for name, value in table.items()}
            value = lower_table.get(str(PHI_MOTOR_NAME).lower())
            values.append(np.nan if value is None else value)
        return np.asarray(values, dtype=float), f"motor:{PHI_MOTOR_NAME}", []

    all_names = sorted(set().union(*[_header_numeric_candidates(header).keys() for header in headers]))
    summaries = []
    for name in all_names:
        values = []
        for header in headers:
            values.append(_header_numeric_candidates(header).get(name, np.nan))
        values = np.asarray(values, dtype=float)
        finite = np.isfinite(values)
        if finite.sum() < max(3, len(values) // 2):
            continue
        value_range = float(np.nanmax(values) - np.nanmin(values))
        summaries.append({
            "name": name,
            "range": value_range,
            "first": float(values[finite][0]),
            "last": float(values[finite][-1]),
            "values": values,
        })

    varying = [item for item in summaries if item["range"] > 1e-12]
    preferred_words = ["phi", "rot", "rock", "omega", "diffrz", "eta"]
    preferred = [
        item for item in varying
        if any(word in item["name"].lower() for word in preferred_words)
    ]
    pool = preferred if preferred else varying
    if not pool:
        return np.full(len(headers), np.nan), None, summaries

    chosen = max(pool, key=lambda item: item["range"])
    return chosen["values"], chosen["name"], summaries


def print_phi_candidate_summary(summaries, top=20):
    if not summaries:
        print("No numeric phi/motor/header candidates found.")
        return
    print("\nNumeric EDF header/motor candidates sorted by range:")
    for item in sorted(summaries, key=lambda x: x["range"], reverse=True)[:top]:
        print(
            f"  {item['name']:28s} range={item['range']:.8g} "
            f"first={item['first']:.8g} last={item['last']:.8g}"
        )


def read_edf_image_and_header(path):
    try:
        import fabio
    except ImportError as exc:
        raise ImportError("Reading EDF headers requires fabio.") from exc

    edf = fabio.open(str(path))
    return edf.data.astype(np.float32), dict(edf.header)


def load_edf_stack_recursive(candidates, angle_deg=18.0, apply_stretch=True, max_frames=None):
    data_dir, files = resolve_data_dir_and_files(candidates)

    images = []
    headers = []
    for path in files:
        image, header = read_edf_image_and_header(path)
        images.append(image)
        headers.append(header)

    phi_values, phi_source, phi_summaries = choose_phi_series(headers)
    print_phi_candidate_summary(phi_summaries)

    records = []
    for path, image, header, phi_value in zip(files, images, headers, phi_values):
        records.append({
            "path": path,
            "image": image,
            "header": header,
            "phi": phi_value,
            "filename_number": filename_numeric_key(path),
        })

    if READ_PHI_FROM_EDF and np.all(np.isfinite(phi_values)) and phi_source is not None:
        records = sorted(records, key=lambda record: record["phi"])
        sort_mode = f"EDF {phi_source}"
    else:
        records = sorted(records, key=lambda record: (record["filename_number"], natural_sort_key(record["path"])))
        sort_mode = "filename numeric order"

    if max_frames is not None:
        records = records[:max_frames]

    frames = []
    for record in records:
        image = record["image"]
        if apply_stretch:
            image = stretch_vertical(image, angle_deg=angle_deg)
        frames.append(image)

    stack = np.stack(frames, axis=0).astype(np.float32)
    sorted_files = [record["path"] for record in records]
    sorted_phi = np.asarray([record["phi"] for record in records], dtype=np.float64)
    return stack, sorted_files, data_dir, sorted_phi, phi_source, sort_mode
```

## Inspect EDF headers for phi

Run this section before plotting if the phi values look wrong. It prints the EDF header fields that contain phi/rotation/motor information and shows the extracted numeric phi values in sorted order.


```python
def inspect_edf_phi_headers(candidates, max_files=8):
    data_dir, files = resolve_data_dir_and_files(candidates)
    print("resolved data dir:", data_dir)
    print("total EDF files:", len(files))

    headers = []
    for path in files:
        _, header = read_edf_image_and_header(path)
        headers.append(header)

    phi_values, phi_source, summaries = choose_phi_series(headers)
    print("\nChosen phi source:", phi_source)
    print("Chosen phi values:", phi_values)
    print_phi_candidate_summary(summaries, top=30)

    for path, header, phi_value in zip(files[:max_files], headers[:max_files], phi_values[:max_files]):
        print("\n" + "=" * 90)
        print("file:", path.name)
        print("chosen phi:", phi_value, "source:", phi_source)
        print("motor table:")
        table = _motor_table_from_header(header)
        if table:
            for name, value in table.items():
                print(f"  {name}: {value}")
        else:
            print("  no motor_mne/motor_pos table found")

    return phi_values, phi_source, summaries


edf_phi_diagnostic, edf_phi_source_diagnostic, phi_candidate_summaries = inspect_edf_phi_headers(
    DATA_DIR_CANDIDATES,
    max_files=8,
)
```

```python
stack, edf_files, resolved_data_dir, edf_phi, edf_phi_source, sort_mode = load_edf_stack_recursive(
    DATA_DIR_CANDIDATES,
    angle_deg=STRETCH_ANGLE_DEG,
    apply_stretch=APPLY_VERTICAL_STRETCH,
)

print("resolved data dir:", resolved_data_dir)
print("number of EDF files:", len(edf_files))
print("sort mode:", sort_mode)
print("chosen phi source:", edf_phi_source)
print("phi values sorted:", edf_phi)
print("loaded stack:", stack.shape)
print("first file:", edf_files[0].name)
print("last file:", edf_files[-1].name)

roi_stack, roi_bounds = crop_stack(
    stack,
    roi_xyxy=ROI_XYXY,
    center_crop_shape=CENTER_CROP_SHAPE,
)

if SUBTRACT_FRAME_MEAN:
    roi_stack_for_curve = subtract_frame_mean(roi_stack)
else:
    roi_stack_for_curve = roi_stack.astype(np.float32)

print("ROI bounds (x0, y0, x1, y1):", roi_bounds)
print("ROI stack:", roi_stack_for_curve.shape)
```

```python
# Integrated ROI intensity per rocking frame.
# Mean intensity keeps the scale independent of ROI size. Use .sum(axis=(1, 2)) if you prefer total counts.
intensity = roi_stack_for_curve.mean(axis=(1, 2))

# Normalize to a visually convenient arbitrary-unit scale like the reference figure.
intensity = intensity - intensity.min()
if intensity.max() > 0:
    intensity = 100.0 + 250.0 * intensity / intensity.max()

num_frames = roi_stack_for_curve.shape[0]

if READ_PHI_FROM_EDF and np.all(np.isfinite(edf_phi)):
    phi_raw = edf_phi.astype(np.float64)
    if PHI_ZERO_MODE == "first":
        phi = (phi_raw - phi_raw[0]) * PHI_PLOT_SCALE
    elif PHI_ZERO_MODE == "min":
        phi = (phi_raw - phi_raw.min()) * PHI_PLOT_SCALE
    else:
        phi = phi_raw * PHI_PLOT_SCALE
else:
    print("Warning: missing EDF phi values; using PHI_START_MDEG/PHI_STEP_MDEG fallback.")
    phi = PHI_START_MDEG + PHI_STEP_MDEG * np.arange(num_frames)

if INSET_FRAMES is None:
    inset_frames = choose_inset_frames(num_frames, HIGHLIGHT_FRAME)
else:
    inset_frames = [int(f) for f in INSET_FRAMES if 0 <= int(f) < num_frames]

highlight_frame = int(np.clip(HIGHLIGHT_FRAME, 0, num_frames - 1))

print("inset frames:", inset_frames)
print("highlight frame:", highlight_frame)
```

```python
plt.rcParams.update({
    "font.size": 14,
    "axes.labelsize": 18,
    "xtick.labelsize": 13,
    "ytick.labelsize": 13,
})

fig, ax = plt.subplots(figsize=(8.3, 6.0))

ax.plot(
    phi,
    intensity,
    color="blue",
    marker="o",
    markersize=6,
    linewidth=1.8,
)
ax.scatter(
    phi[highlight_frame],
    intensity[highlight_frame],
    color="red",
    s=55,
    zorder=5,
)

ax.set_xlabel(PHI_XLABEL)
ax.set_ylabel("Intensity [A.U.]")
ax.grid(True, linestyle="--", alpha=0.45)

# Leave room above the curve for image insets.
y_min = float(np.min(intensity))
y_max = float(np.max(intensity))
y_range = max(1.0, y_max - y_min)
ax.set_ylim(y_min - 0.10 * y_range, y_max + 0.30 * y_range)
ax.set_xlim(phi.min() - 0.8, phi.max() + 0.8)

# Place each inset near the corresponding curve point, alternating vertical offsets.
offset_pattern = [72, 96, 74, 56, 36, 18, 34, 58, 80]
for k, frame in enumerate(inset_frames):
    image = prepare_inset_image(roi_stack_for_curve[frame])
    offset_y = offset_pattern[k % len(offset_pattern)]
    offset_x = 0 if k % 2 == 0 else 16

    image_box = OffsetImage(image, cmap=CMAP, zoom=INSET_ZOOM)
    edgecolor = "red" if frame == highlight_frame else "black"
    linewidth = 3.0 if frame == highlight_frame else 0.8

    annotation = AnnotationBbox(
        image_box,
        (phi[frame], intensity[frame]),
        xybox=(offset_x, offset_y),
        xycoords="data",
        boxcoords="offset points",
        frameon=True,
        pad=0.0,
        bboxprops={"edgecolor": edgecolor, "linewidth": linewidth, "facecolor": "none"},
        arrowprops=None,
        zorder=4,
    )
    ax.add_artist(annotation)

# Red arrow from the weak-beam marker to the highlighted inset.
ax.annotate(
    "",
    xy=(phi[highlight_frame], intensity[highlight_frame]),
    xytext=(phi[highlight_frame] - 2.2, intensity[highlight_frame] - 0.18 * y_range),
    arrowprops={"arrowstyle": "-|>", "color": "red", "lw": 1.8},
    zorder=6,
)

fig.savefig(OUTPUT_PNG, dpi=300, bbox_inches="tight")
fig.savefig(OUTPUT_PDF, dpi=300, bbox_inches="tight")
print("saved:", OUTPUT_PNG.resolve())
print("saved:", OUTPUT_PDF.resolve())
plt.show()
```

## Notes

- By default the notebook uses a centered `1200 x 1200` ROI through `CENTER_CROP_SHAPE`. Set `ROI_XYXY` if you want a specific region instead.
- Adjust `PHI_START_MDEG` and `PHI_STEP_MDEG` to match your rocking scan calibration.
- Adjust `HIGHLIGHT_FRAME` to mark the weak-beam image you want to emphasize.
- If the insets overlap too much, reduce `INSET_ZOOM`, reduce `INSET_MAX_PIXELS`, or manually set `INSET_FRAMES`.
- `SUBTRACT_FRAME_MEAN = False`, so the plot uses raw loaded/stretched intensities without background subtraction.
