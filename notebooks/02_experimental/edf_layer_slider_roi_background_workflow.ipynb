# Layer Slider ROI Selection and Background-Corrected Integration

This notebook shows the 4th rocking image from each layer in a layer slider. Select one ROI for each layer. Then it subtracts the background level `mean + std` from each frame, integrates from the 4th frame to the end of that layer's rocking curve, resizes to the neural-network input size `(510, 170)`, and saves each integrated image with its layer number.

```python
from pathlib import Path

try:
    get_ipython().run_line_magic("matplotlib", "widget")
except Exception:
    get_ipython().run_line_magic("matplotlib", "notebook")

import json
import matplotlib.pyplot as plt
import numpy as np

from bv_repro.data import (
    LayerFrameROISelector,
    load_edf_stack,
    load_layer_frame_preview,
    save_layer_integrated_from_roi,
    subtract_full_image_mean,
)
```

## Configure

```python
ROOT_DIR = Path(
    "/mnt/multipath-shares/data/projects/engage_id03/cross_slip_data/"
    "Al_sample_5_tick_revolution_70/Al_sample_5_tick_revolution_70"
)

if not ROOT_DIR.exists():
    ROOT_DIR = Path(
        "/data/projects/engage_id03/cross_slip_data/"
        "Al_sample_5_tick_revolution_70/Al_sample_5_tick_revolution_70"
    )

LAYER_PATTERN = "Z_local_rocking_2x_*"
OUTPUT_DIR = Path("edf_layer_background_corrected_integrated")

STRETCH_ANGLE_DEG = 18.0
START_FRAME = 3  # zero-based: 3 means the 4th rocking image
TARGET_SHAPE = (510, 170)  # neural-network input size: H, W
CENTER_CROP_SHAPE = (1000, 1000)  # centered crop shown for ROI selection: H, W
CLIP_NEGATIVE_AFTER_BACKGROUND = True

DISPLAY_CMAP = "viridis"
DISPLAY_LOG_SCALE = True
DISPLAY_TRANSPOSE = True
DISPLAY_PERCENTILES = (1.0, 99.5)  # contrast percentiles for visualization only

print("ROOT_DIR:", ROOT_DIR)
print("exists:", ROOT_DIR.exists())
```

## Load and background-correct the rocking stacks

For each layer, this loads the full rocking stack, subtracts the mean from every full image/frame, clips negatives to zero, and uses the corrected 4th image for ROI selection. The later integration uses these already-corrected stacks without further background subtraction.


```python
layer_dirs = sorted([p for p in ROOT_DIR.glob(LAYER_PATTERN) if p.is_dir()])
if not layer_dirs:
    raise FileNotFoundError(f"No layer directories matching {LAYER_PATTERN!r} in {ROOT_DIR}")

corrected_stacks = {}
background_levels_by_layer = {}
layer_previews = []

for layer_dir in layer_dirs:
    print("Loading and correcting:", layer_dir.name)
    stack, files = load_edf_stack(
        layer_dir,
        angle_deg=STRETCH_ANGLE_DEG,
        apply_stretch=True,
    )
    corrected_stack, background_levels = subtract_full_image_mean(
        stack,
        clip_negative=CLIP_NEGATIVE_AFTER_BACKGROUND,
    )

    corrected_stacks[layer_dir.name] = corrected_stack
    background_levels_by_layer[layer_dir.name] = background_levels
    layer_previews.append({
        "layer_dir": layer_dir,
        "file": files[START_FRAME],
        "image": corrected_stack[START_FRAME],
    })

for i, preview in enumerate(layer_previews):
    print(f"{i:02d}: {preview['layer_dir'].name} -> {preview['file'].name}")

print("number of layers:", len(layer_previews))
```

## Select one ROI per layer

Move the layer slider. For each layer, draw a rectangle ROI on the 4th rocking image. The selector stores one ROI per layer, so you should end with the same number of ROIs as layers.

```python
selector = LayerFrameROISelector(
    layer_previews,
    cmap=DISPLAY_CMAP,
    log_scale=DISPLAY_LOG_SCALE,
    transpose=DISPLAY_TRANSPOSE,
    crop_center_shape=CENTER_CROP_SHAPE,
    display_percentiles=DISPLAY_PERCENTILES,
)
selector.show()
```

## Check selected ROIs

```python
rois = selector.get_rois()
print("selected ROIs:", len(rois), "/", len(layer_previews))

missing = [p["layer_dir"].name for p in layer_previews if p["layer_dir"].name not in rois]
if missing:
    print("Missing ROI for layers:")
    for name in missing:
        print("  ", name)
else:
    print("All layers have an ROI.")

for name, roi in rois.items():
    print(name, roi)
```

## Integrate all selected layer ROIs

For every selected layer, this crops the ROI from the already mean-corrected rocking stack and integrates from the 4th frame to the end. No additional background subtraction is applied during integration.


```python
records = []
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

for preview in layer_previews:
    layer_dir = preview["layer_dir"]
    layer_name = layer_dir.name
    if layer_name not in rois:
        print("Skipping, no ROI:", layer_name)
        continue

    print("Integrating corrected stack:", layer_name)
    corrected_stack = corrected_stacks[layer_name]

    outputs, metadata = save_layer_integrated_from_roi(
        corrected_stack,
        roi=rois[layer_name],
        output_dir=OUTPUT_DIR,
        layer_name=layer_name,
        start_frame=START_FRAME,
        target_shape=TARGET_SHAPE,
        background_subtract=False,
        clip_negative=False,
        save_corrected_stack=False,
    )

    np.save(OUTPUT_DIR / f"layer_{metadata['layer_number']}_full_image_background_levels.npy", background_levels_by_layer[layer_name])
    metadata["pre_correction"] = "full image per frame: frame - mean"
    metadata["clip_negative_after_pre_correction"] = bool(CLIP_NEGATIVE_AFTER_BACKGROUND)

    with (OUTPUT_DIR / f"layer_{metadata['layer_number']}_metadata.json").open("w", encoding="utf-8") as f:
        json.dump(metadata, f, indent=2)

    records.append({
        "layer_name": layer_name,
        "metadata": metadata,
        "image": outputs["integrated_image"],
    })

print("processed layers:", len(records))
```

## Save combined neural-network input stack and preview

```python
if records:
    nn_stack = np.stack([r["image"] for r in records], axis=0).astype(np.float32)
    layer_names = [r["layer_name"] for r in records]

    np.save(OUTPUT_DIR / "all_layers_integrated_nn_input.npy", nn_stack)
    with (OUTPUT_DIR / "all_layers_integrated_nn_input_metadata.json").open("w", encoding="utf-8") as f:
        json.dump({"layer_names": layer_names, "shape": list(nn_stack.shape)}, f, indent=2)

    print("combined stack shape:", nn_stack.shape)
    print("saved:", (OUTPUT_DIR / "all_layers_integrated_nn_input.npy").resolve())

    cols = min(4, len(records))
    rows = int(np.ceil(len(records) / cols))
    fig, axes = plt.subplots(rows, cols, figsize=(4 * cols, 4 * rows), squeeze=False)
    axes = axes.ravel()

    for ax, record in zip(axes, records):
        ax.imshow(record["image"], cmap="viridis")
        ax.set_title(record["layer_name"])
        ax.set_axis_off()

    for ax in axes[len(records):]:
        ax.set_axis_off()

    plt.tight_layout()
    plt.savefig(OUTPUT_DIR / "all_layers_integrated_preview.png", dpi=300, bbox_inches="tight")
    plt.show()
else:
    print("No records to save. Select ROIs and run integration first.")
```
