# Mean-Corrected ROI Picking, Integration, Crops, and Video

This notebook does the full workflow in one place:

1. Load each layer rocking stack.
2. Subtract the full-frame mean from every frame.
3. Show the corrected 4th image from each layer as a centered `1000 x 1000` view.
4. Pick one dislocation ROI per layer.
5. Save the ROI integrated from the 4th frame to the end.
6. Show/save the non-integrated ROI crop from the corrected 4th image.
7. Make a video of the centered `1000 x 1000` corrected 4th images with thin bounding boxes.

```python
from pathlib import Path
import json

try:
    get_ipython().run_line_magic("matplotlib", "widget")
except Exception:
    get_ipython().run_line_magic("matplotlib", "notebook")

import imageio.v2 as imageio
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Rectangle

from bv_repro.data import (
    LayerFrameROISelector,
    load_edf_stack,
    save_layer_integrated_from_roi,
    subtract_full_image_mean,
)
```

## Configure

```python
ROOT_DIR = Path(
    "/mnt/multipath-shares/data/projects/engage_id03/cross_slip_data/"
    "Al_sample_5_tick_revolution_70/Al_sample_5_tick_revolution_70"
)

if not ROOT_DIR.exists():
    ROOT_DIR = Path(
        "/data/projects/engage_id03/cross_slip_data/"
        "Al_sample_5_tick_revolution_70/Al_sample_5_tick_revolution_70"
    )

LAYER_PATTERN = "Z_local_rocking_2x_*"
OUTPUT_DIR = Path("roi_mean_correct_integrate_and_video_outputs")

STRETCH_ANGLE_DEG = 18.0
START_FRAME = 3  # zero-based: 3 means 4th rocking image
CENTER_CROP_SHAPE = (1000, 1000)
TARGET_SHAPE = (510, 170)  # neural-network input size H, W

DISPLAY_CMAP = "viridis"
DISPLAY_LOG_SCALE = True
DISPLAY_TRANSPOSE = False
DISPLAY_PERCENTILES = (1.0, 99.5)

PADDING_PX = 20
VIDEO_FPS = 2
BBOX_LINEWIDTH = 1.0

PIXEL_SIZE_HORIZONTAL_UM = 0.017
SCALE_BAR_UM = 5.0

OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

print("ROOT_DIR:", ROOT_DIR)
print("exists:", ROOT_DIR.exists())
```

## Load stacks and subtract mean from every frame

This step subtracts `frame.mean()` from each full image before ROI selection or integration.

```python
layer_dirs = sorted([p for p in ROOT_DIR.glob(LAYER_PATTERN) if p.is_dir()])
if not layer_dirs:
    raise FileNotFoundError(f"No layer directories matching {LAYER_PATTERN!r} in {ROOT_DIR}")

corrected_stacks = {}
background_means_by_layer = {}
layer_previews = []

for layer_dir in layer_dirs:
    print("Loading and mean-correcting:", layer_dir.name)
    stack, files = load_edf_stack(
        layer_dir,
        angle_deg=STRETCH_ANGLE_DEG,
        apply_stretch=True,
    )
    corrected_stack, frame_means = subtract_full_image_mean(stack, clip_negative=True)
    corrected_stacks[layer_dir.name] = corrected_stack
    background_means_by_layer[layer_dir.name] = frame_means
    layer_previews.append({
        "layer_dir": layer_dir,
        "file": files[START_FRAME],
        "image": corrected_stack[START_FRAME],
    })

for i, preview in enumerate(layer_previews):
    print(f"{i:02d}: {preview['layer_dir'].name} -> {preview['file'].name}")

print("number of layers:", len(layer_previews))
```

## Pick one ROI per layer on the centered 1000 x 1000 corrected image

```python
selector = LayerFrameROISelector(
    layer_previews,
    cmap=DISPLAY_CMAP,
    log_scale=DISPLAY_LOG_SCALE,
    transpose=DISPLAY_TRANSPOSE,
    crop_center_shape=CENTER_CROP_SHAPE,
    display_percentiles=DISPLAY_PERCENTILES,
)
selector.show()
```

## Integrate selected ROIs from the 4th frame to the end

No further background subtraction is done here because the stacks were already mean-corrected above.

```python
rois = selector.get_rois()
print("selected ROIs:", len(rois), "/", len(layer_previews))

integration_records = []

for preview in layer_previews:
    layer_name = preview["layer_dir"].name
    if layer_name not in rois:
        print("Skipping integration, no ROI:", layer_name)
        continue

    corrected_stack = corrected_stacks[layer_name]
    outputs, metadata = save_layer_integrated_from_roi(
        corrected_stack,
        roi=rois[layer_name],
        output_dir=OUTPUT_DIR,
        layer_name=layer_name,
        start_frame=START_FRAME,
        target_shape=TARGET_SHAPE,
        background_subtract=False,
        clip_negative=False,
        save_corrected_stack=False,
    )

    np.save(OUTPUT_DIR / f"layer_{metadata['layer_number']}_frame_means.npy", background_means_by_layer[layer_name])
    metadata["pre_correction"] = "full image per frame: frame - mean"
    metadata["integrated_from_frame_zero_based"] = START_FRAME

    with (OUTPUT_DIR / f"layer_{metadata['layer_number']}_metadata.json").open("w", encoding="utf-8") as f:
        json.dump(metadata, f, indent=2)

    integration_records.append({
        "layer_name": layer_name,
        "metadata": metadata,
        "integrated_image": outputs["integrated_image"],
    })

print("integrated layers:", len(integration_records))
```

## Show and save non-integrated ROI crops from the corrected 4th image

```python
def crop_with_padding(image, roi_xyxy, padding=20):
    h, w = image.shape
    x0, y0, x1, y1 = roi_xyxy
    x0 = max(0, int(x0) - padding)
    y0 = max(0, int(y0) - padding)
    x1 = min(w, int(x1) + padding)
    y1 = min(h, int(y1) + padding)
    return image[y0:y1, x0:x1], (x0, y0, x1, y1)


def prepare_display(crop):
    crop = crop.astype(np.float32)
    if DISPLAY_LOG_SCALE:
        crop = np.log1p(np.clip(crop, a_min=0.0, a_max=None))
    vmin, vmax = np.percentile(crop, DISPLAY_PERCENTILES)
    return np.clip(crop, vmin, vmax), vmin, vmax


def add_scale_bar(ax, image_shape):
    h, w = image_shape
    bar_px = min(SCALE_BAR_UM / PIXEL_SIZE_HORIZONTAL_UM, w * 0.55)
    x0 = w * 0.08
    y0 = h * 0.90
    x1 = x0 + bar_px
    ax.plot([x0, x1], [y0, y0], color="red", linewidth=3.0, solid_capstyle="butt")
    ax.text((x0 + x1) / 2, y0 - h * 0.045, f"{SCALE_BAR_UM:g} Âµm", color="white", ha="center", va="bottom", fontsize=11, fontweight="bold")


crop_records = []

for preview in layer_previews:
    layer_name = preview["layer_dir"].name
    if layer_name not in rois:
        print("Skipping crop, no ROI:", layer_name)
        continue

    image = corrected_stacks[layer_name][START_FRAME]
    crop, crop_bounds = crop_with_padding(image, rois[layer_name], padding=PADDING_PX)
    display_crop, vmin, vmax = prepare_display(crop)

    fig, ax = plt.subplots(figsize=(5, 5))
    ax.imshow(display_crop, cmap=DISPLAY_CMAP, vmin=vmin, vmax=vmax)
    add_scale_bar(ax, display_crop.shape)
    ax.set_title(layer_name)
    ax.set_axis_off()
    plt.tight_layout(pad=0.1)

    out_png = OUTPUT_DIR / f"{layer_name}_corrected_4th_roi_crop.png"
    out_npy = OUTPUT_DIR / f"{layer_name}_corrected_4th_roi_crop.npy"
    plt.savefig(out_png, dpi=300, bbox_inches="tight", pad_inches=0.02)
    plt.show()

    np.save(out_npy, crop.astype(np.float32))
    crop_records.append({
        "layer_name": layer_name,
        "roi_xyxy": list(map(int, rois[layer_name])),
        "crop_bounds_xyxy": list(map(int, crop_bounds)),
        "crop_shape": list(map(int, crop.shape)),
        "png": str(out_png),
        "npy": str(out_npy),
    })

with (OUTPUT_DIR / "corrected_4th_roi_crop_metadata.json").open("w", encoding="utf-8") as f:
    json.dump(crop_records, f, indent=2)

print("saved non-integrated crops:", len(crop_records))
```

## Make video of centered 1000 x 1000 corrected images with thin bounding boxes

```python
def center_crop_bounds(image_shape, crop_shape):
    h, w = image_shape
    crop_h, crop_w = crop_shape
    crop_h = min(int(crop_h), h)
    crop_w = min(int(crop_w), w)
    y0 = max(0, (h - crop_h) // 2)
    x0 = max(0, (w - crop_w) // 2)
    return x0, y0, x0 + crop_w, y0 + crop_h


frames = []

for preview in layer_previews:
    layer_name = preview["layer_dir"].name
    image = corrected_stacks[layer_name][START_FRAME]
    cx0, cy0, cx1, cy1 = center_crop_bounds(image.shape, CENTER_CROP_SHAPE)
    center_crop = image[cy0:cy1, cx0:cx1]
    display_img, vmin, vmax = prepare_display(center_crop)

    fig, ax = plt.subplots(figsize=(6, 6), dpi=120)
    ax.imshow(display_img, cmap=DISPLAY_CMAP, vmin=vmin, vmax=vmax)
    ax.set_title(layer_name)
    ax.set_axis_off()

    if layer_name in rois:
        x0, y0, x1, y1 = rois[layer_name]
        dx0 = x0 - cx0
        dx1 = x1 - cx0
        dy0 = y0 - cy0
        dy1 = y1 - cy0
        ax.add_patch(
            Rectangle(
                (dx0, dy0),
                dx1 - dx0,
                dy1 - dy0,
                fill=False,
                edgecolor="red",
                linewidth=BBOX_LINEWIDTH,
            )
        )

    fig.canvas.draw()
    frame = np.asarray(fig.canvas.buffer_rgba())[:, :, :3].copy()
    frames.append(frame)
    plt.close(fig)

video_path = OUTPUT_DIR / "corrected_center_1000_bbox_video.gif"
imageio.mimsave(video_path, frames, duration=1 / VIDEO_FPS)
print("saved video:", video_path.resolve())
```
