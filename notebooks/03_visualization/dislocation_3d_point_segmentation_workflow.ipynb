# Dislocation Point Segmentation Through Layers

Use this notebook after creating the layer-integrated images. It shows one layer image at a time with a slider. Click the dislocation point in each layer, then convert the selected pixel positions to lab coordinates and plot the dislocation line in 3D.

```python
from pathlib import Path

try:
    get_ipython().run_line_magic("matplotlib", "widget")
except Exception:
    get_ipython().run_line_magic("matplotlib", "notebook")

import matplotlib.pyplot as plt
import numpy as np

from bv_repro.dislocation_point_tools import (
    LayerPointSelector,
    load_integrated_layer_images,
    plot_dislocation_3d,
    points_to_xyz,
    save_points,
)
```

## Configure

Set the pixel size and layer spacing in microns. If you do not know them yet, keep `1.0` and the plot will be in pixel-equivalent units.

```python
INPUT_DIR = Path("edf_layer_background_corrected_integrated")
OUTPUT_DIR = Path("dislocation_3d_points")

# Paper geometry calibration.
# Detector image: 2160 x 2560 px, effective lab-frame pixel size 57 x 17 nm.
# Field of view: 123 um x 44 um. 11 layers spaced by 2 um -> 20 um z span.
PIXEL_SIZE_X_UM = 0.057
PIXEL_SIZE_Y_UM = 0.017
LAYER_STEP_Z_UM = 2.0

VOLUME_X_UM = 2160 * PIXEL_SIZE_X_UM
VOLUME_Y_UM = 2560 * PIXEL_SIZE_Y_UM
VOLUME_Z_UM = 10 * LAYER_STEP_Z_UM

ORIGIN_X_UM = 0.0
ORIGIN_Y_UM = 0.0
ORIGIN_Z_UM = 0.0

DISPLAY_CMAP = "viridis"
DISPLAY_LOG_SCALE = False
DISPLAY_TRANSPOSE = False

print("INPUT_DIR:", INPUT_DIR.resolve())
```

## Load layer images

```python
images, layer_names = load_integrated_layer_images(INPUT_DIR)

print("images shape:", images.shape)
print("number of layers:", len(layer_names))
for i, name in enumerate(layer_names):
    print(f"{i:02d}: {name}")
```

## Click one dislocation point per layer

Move the layer slider. Click the dislocation point in each layer. If you click again on the same layer, the point is replaced.

```python
selector = LayerPointSelector(
    images,
    layer_names=layer_names,
    cmap=DISPLAY_CMAP,
    log_scale=DISPLAY_LOG_SCALE,
    transpose=DISPLAY_TRANSPOSE,
)
selector.show()
```

## Convert selected points to 3D coordinates

```python
points = selector.get_points()
print("selected points:", len(points), "/", len(layer_names))

missing = [name for name in layer_names if name not in points]
if missing:
    print("Missing points for layers:")
    for name in missing:
        print("  ", name)

rows = points_to_xyz(
    points,
    layer_names,
    pixel_size_x_um=PIXEL_SIZE_X_UM,
    pixel_size_y_um=PIXEL_SIZE_Y_UM,
    layer_step_z_um=LAYER_STEP_Z_UM,
    origin_x_um=ORIGIN_X_UM,
    origin_y_um=ORIGIN_Y_UM,
    origin_z_um=ORIGIN_Z_UM,
)

for row in rows:
    print(
        f"{row['layer_name']}: "
        f"x={row['x_um']:.3f} um, y={row['y_um']:.3f} um, z={row['z_um']:.3f} um"
    )
```

## Save points and plot in 3D

```python
json_path, csv_path = save_points(rows, OUTPUT_DIR)
print("saved JSON:", json_path.resolve())
print("saved CSV:", csv_path.resolve())

plot_dislocation_3d(
    rows,
    output_path=OUTPUT_DIR / "dislocation_trajectory_3d.png",
    title="3D Dislocation Line",
    xlim=(0, VOLUME_X_UM),
    ylim=(0, VOLUME_Y_UM),
    zlim=(0, VOLUME_Z_UM),
)
```
