# Saved ROI Crops With Scale Bar

This notebook reloads the saved dislocation bounding boxes, crops those ROIs from the 4th stretched EDF image of each layer, does **no integration**, does **no transpose**, and displays/saves the cropped ROI images with a micrometer scale bar.

```python
from pathlib import Path
import json

import matplotlib.pyplot as plt
import numpy as np

from bv_repro.data import load_layer_frame_preview
```

## Configure

```python
ROOT_DIR = Path(
    "/mnt/multipath-shares/data/projects/engage_id03/cross_slip_data/"
    "Al_sample_5_tick_revolution_70/Al_sample_5_tick_revolution_70"
)

if not ROOT_DIR.exists():
    ROOT_DIR = Path(
        "/data/projects/engage_id03/cross_slip_data/"
        "Al_sample_5_tick_revolution_70/Al_sample_5_tick_revolution_70"
    )

BOXES_JSON = Path("bbox_segmentation_video_outputs/segmented_bounding_boxes.json")
OUTPUT_DIR = Path("saved_roi_crops_no_transpose_scalebar")

LAYER_PATTERN = "Z_local_rocking_2x_*"
FRAME_INDEX = 3  # 4th rocking image
STRETCH_ANGLE_DEG = 18.0

# No transpose: horizontal axis is detector-column direction.
PIXEL_SIZE_HORIZONTAL_UM = 0.017
SCALE_BAR_UM = 5.0

DISPLAY_CMAP = "viridis"
DISPLAY_LOG_SCALE = True
DISPLAY_PERCENTILES = (1.0, 99.5)
PADDING_PX = 20

OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

print("ROOT_DIR:", ROOT_DIR)
print("BOXES_JSON:", BOXES_JSON)
```

## Load boxes and images

```python
with BOXES_JSON.open("r", encoding="utf-8") as f:
    box_rows = json.load(f)

boxes = {row["layer_name"]: tuple(row["bbox_xyxy"]) for row in box_rows}
print("loaded boxes:", len(boxes))

layer_previews = load_layer_frame_preview(
    ROOT_DIR,
    frame_index=FRAME_INDEX,
    pattern=LAYER_PATTERN,
    angle_deg=STRETCH_ANGLE_DEG,
    apply_stretch=True,
)

print("loaded layer images:", len(layer_previews))
```

## Crop, display, and save

```python
def crop_with_padding(image, bbox_xyxy, padding=20):
    height, width = image.shape
    x0, y0, x1, y1 = bbox_xyxy
    x0 = max(0, int(x0) - padding)
    y0 = max(0, int(y0) - padding)
    x1 = min(width, int(x1) + padding)
    y1 = min(height, int(y1) + padding)
    return image[y0:y1, x0:x1], (x0, y0, x1, y1)


def prepare_display(crop):
    crop = crop.astype(np.float32)
    if DISPLAY_LOG_SCALE:
        crop = np.log1p(np.clip(crop, a_min=0.0, a_max=None))
    vmin, vmax = np.percentile(crop, DISPLAY_PERCENTILES)
    return np.clip(crop, vmin, vmax), vmin, vmax


def add_scale_bar(ax, image_shape, length_um, pixel_size_um):
    height, width = image_shape
    bar_px = length_um / pixel_size_um
    bar_px = min(bar_px, width * 0.55)
    x0 = width * 0.08
    y0 = height * 0.90
    x1 = x0 + bar_px

    ax.plot([x0, x1], [y0, y0], color="red", linewidth=3.5, solid_capstyle="butt")
    ax.text(
        (x0 + x1) / 2,
        y0 - height * 0.045,
        f"{length_um:g} Âµm",
        color="white",
        ha="center",
        va="bottom",
        fontsize=12,
        fontweight="bold",
    )


metadata = []

for preview in layer_previews:
    layer_name = preview["layer_dir"].name
    if layer_name not in boxes:
        continue

    image = preview["image"].astype(np.float32)
    crop, crop_bounds = crop_with_padding(image, boxes[layer_name], padding=PADDING_PX)
    display_crop, vmin, vmax = prepare_display(crop)

    fig, ax = plt.subplots(figsize=(5, 5))
    ax.imshow(display_crop, cmap=DISPLAY_CMAP, vmin=vmin, vmax=vmax)
    add_scale_bar(ax, display_crop.shape, SCALE_BAR_UM, PIXEL_SIZE_HORIZONTAL_UM)
    ax.set_title(layer_name)
    ax.set_axis_off()
    plt.tight_layout(pad=0.1)

    out_png = OUTPUT_DIR / f"{layer_name}_roi_crop_scalebar.png"
    out_npy = OUTPUT_DIR / f"{layer_name}_roi_crop.npy"
    plt.savefig(out_png, dpi=300, bbox_inches="tight", pad_inches=0.02)
    plt.show()

    np.save(out_npy, crop.astype(np.float32))
    metadata.append({
        "layer_name": layer_name,
        "bbox_xyxy": list(map(int, boxes[layer_name])),
        "crop_bounds_xyxy": list(map(int, crop_bounds)),
        "crop_shape": list(map(int, crop.shape)),
        "png": str(out_png),
        "npy": str(out_npy),
        "scale_bar_um": SCALE_BAR_UM,
        "pixel_size_horizontal_um": PIXEL_SIZE_HORIZONTAL_UM,
        "transposed": False,
        "integrated": False,
    })

with (OUTPUT_DIR / "roi_crop_metadata.json").open("w", encoding="utf-8") as f:
    json.dump(metadata, f, indent=2)

print("saved crops:", len(metadata))
print("output dir:", OUTPUT_DIR.resolve())
```
