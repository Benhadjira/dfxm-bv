# Plot Saved Dislocation ROI Crops With Scale Bars

This notebook loads the saved bounding boxes from the dislocation segmentation workflow, reloads the 4th stretched EDF image from each layer, crops the saved ROI, and saves publication-style ROI crop images with a micrometer scale bar.

```python
from pathlib import Path
import json
import math

import matplotlib.pyplot as plt
import numpy as np

from bv_repro.data import load_layer_frame_preview
```

## Configure

```python
ROOT_DIR = Path(
    "/mnt/multipath-shares/data/projects/engage_id03/cross_slip_data/"
    "Al_sample_5_tick_revolution_70/Al_sample_5_tick_revolution_70"
)

if not ROOT_DIR.exists():
    ROOT_DIR = Path(
        "/data/projects/engage_id03/cross_slip_data/"
        "Al_sample_5_tick_revolution_70/Al_sample_5_tick_revolution_70"
    )

BOXES_JSON = Path("bbox_segmentation_video_outputs/segmented_bounding_boxes.json")
OUTPUT_DIR = Path("dislocation_roi_crops_with_scalebar")

LAYER_PATTERN = "Z_local_rocking_2x_*"
FRAME_INDEX = 3  # 4th rocking image
STRETCH_ANGLE_DEG = 18.0

# Paper geometry: detector columns correspond to lab y with 17 nm pixel size.
# The scale bar is horizontal in the displayed, untransposed crop.
PIXEL_SIZE_HORIZONTAL_UM = 0.017
SCALE_BAR_UM = 5.0

DISPLAY_CMAP = "viridis"
DISPLAY_LOG_SCALE = True
PADDING_PX = 20

OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

print("ROOT_DIR:", ROOT_DIR)
print("BOXES_JSON:", BOXES_JSON)
```

## Load saved boxes and 4th layer images

```python
with BOXES_JSON.open("r", encoding="utf-8") as f:
    rows = json.load(f)

boxes = {row["layer_name"]: tuple(row["bbox_xyxy"]) for row in rows}
print("loaded boxes:", len(boxes))

layer_previews = load_layer_frame_preview(
    ROOT_DIR,
    frame_index=FRAME_INDEX,
    pattern=LAYER_PATTERN,
    angle_deg=STRETCH_ANGLE_DEG,
    apply_stretch=True,
)

print("loaded layer images:", len(layer_previews))
```

## Helper functions

```python
def prepare_crop_for_display(crop, log_scale=True):
    crop = crop.astype(np.float32)
    if log_scale:
        crop = np.log1p(np.clip(crop, a_min=0.0, a_max=None))
    vmin, vmax = np.percentile(crop, [1.0, 99.5])
    return np.clip(crop, vmin, vmax), vmin, vmax


def crop_with_padding(image, bbox_xyxy, padding=20):
    height, width = image.shape
    x0, y0, x1, y1 = bbox_xyxy
    x0 = max(0, int(x0) - padding)
    y0 = max(0, int(y0) - padding)
    x1 = min(width, int(x1) + padding)
    y1 = min(height, int(y1) + padding)
    return image[y0:y1, x0:x1], (x0, y0, x1, y1)


def add_scale_bar(ax, image_shape, length_um=5.0, pixel_size_um=0.017, color="white"):
    height, width = image_shape
    bar_px = length_um / pixel_size_um
    bar_px = min(bar_px, width * 0.55)

    x0 = width * 0.08
    y0 = height * 0.90
    x1 = x0 + bar_px

    ax.plot([x0, x1], [y0, y0], color=color, linewidth=4, solid_capstyle="butt")
    ax.text(
        (x0 + x1) / 2,
        y0 - height * 0.045,
        f"{length_um:g} µm",
        color=color,
        ha="center",
        va="bottom",
        fontsize=12,
        fontweight="bold",
    )
```

## Plot and save ROI crops with scale bars

```python
saved = []

for preview in layer_previews:
    layer_name = preview["layer_dir"].name
    if layer_name not in boxes:
        continue

    image = preview["image"]
    crop, crop_bounds = crop_with_padding(image, boxes[layer_name], padding=PADDING_PX)
    display_crop, vmin, vmax = prepare_crop_for_display(crop, log_scale=DISPLAY_LOG_SCALE)

    fig, ax = plt.subplots(figsize=(5, 5))
    ax.imshow(display_crop, cmap=DISPLAY_CMAP, vmin=vmin, vmax=vmax)
    add_scale_bar(
        ax,
        display_crop.shape,
        length_um=SCALE_BAR_UM,
        pixel_size_um=PIXEL_SIZE_HORIZONTAL_UM,
        color="white",
    )
    ax.set_title(layer_name)
    ax.set_axis_off()
    plt.tight_layout(pad=0.1)

    out_png = OUTPUT_DIR / f"{layer_name}_roi_crop_scalebar.png"
    out_npy = OUTPUT_DIR / f"{layer_name}_roi_crop.npy"
    plt.savefig(out_png, dpi=300, bbox_inches="tight", pad_inches=0.02)
    plt.show()

    np.save(out_npy, crop.astype(np.float32))
    saved.append({
        "layer_name": layer_name,
        "png": str(out_png),
        "npy": str(out_npy),
        "crop_bounds_xyxy": list(map(int, crop_bounds)),
        "bbox_xyxy": list(map(int, boxes[layer_name])),
        "scale_bar_um": SCALE_BAR_UM,
        "pixel_size_horizontal_um": PIXEL_SIZE_HORIZONTAL_UM,
    })

with (OUTPUT_DIR / "roi_crop_scalebar_metadata.json").open("w", encoding="utf-8") as f:
    json.dump(saved, f, indent=2)

print("saved crops:", len(saved))
print("output dir:", OUTPUT_DIR.resolve())
```

## Annotated ROI crop figure

Use this cell to make a publication-style annotated crop like the example: solid/slanted guide line, dotted guide lines, ellipse around the dislocation, and a red scale bar. Edit `LAYER_TO_PLOT` and the annotation parameters until the overlay matches your feature.

```python
from matplotlib.patches import Ellipse

# Pick one saved bbox/layer to annotate.
# Example: "Z_local_rocking_2x_06". If None, use the first saved layer.
LAYER_TO_PLOT = None

# Annotation positions are in crop-display pixel coordinates after padding.
# Tune these for your selected ROI crop.
ANNOTATION = {
    "ellipse_xy": (90, 115),       # center x, y
    "ellipse_width": 45,
    "ellipse_height": 95,
    "ellipse_angle": -22,
    "solid_line": ((25, 245), (185, 0)),      # (x0, y0), (x1, y1)
    "dotted_line": ((0, 20), (160, 270)),     # (x0, y0), (x1, y1)
    "horizontal_dotted_y": 145,
}

SCALE_BAR_UM = 11.5
SCALE_BAR_COLOR = "red"
GUIDE_COLOR = "white"
ELLIPSE_COLOR = "lightgray"

selected = saved[0] if LAYER_TO_PLOT is None else next(s for s in saved if s["layer_name"] == LAYER_TO_PLOT)
layer_name = selected["layer_name"]
crop = np.load(selected["npy"]).astype(np.float32)
display_crop, vmin, vmax = prepare_crop_for_display(crop, log_scale=DISPLAY_LOG_SCALE)

fig, ax = plt.subplots(figsize=(5.2, 6.2))
ax.imshow(display_crop, cmap=DISPLAY_CMAP, vmin=vmin, vmax=vmax)

# Guide lines
(x0, y0), (x1, y1) = ANNOTATION["solid_line"]
ax.plot([x0, x1], [y0, y1], color=GUIDE_COLOR, linewidth=1.8, alpha=0.95)

(x0, y0), (x1, y1) = ANNOTATION["dotted_line"]
ax.plot([x0, x1], [y0, y1], color=GUIDE_COLOR, linewidth=1.5, linestyle=":", alpha=0.85)

yh = ANNOTATION["horizontal_dotted_y"]
ax.axhline(yh, color=GUIDE_COLOR, linewidth=1.5, linestyle=":", alpha=0.85)

# Ellipse around the dislocation
ellipse = Ellipse(
    ANNOTATION["ellipse_xy"],
    width=ANNOTATION["ellipse_width"],
    height=ANNOTATION["ellipse_height"],
    angle=ANNOTATION["ellipse_angle"],
    fill=False,
    edgecolor=ELLIPSE_COLOR,
    linewidth=3.0,
)
ax.add_patch(ellipse)

# Red scale bar
height, width = display_crop.shape
bar_px = SCALE_BAR_UM / PIXEL_SIZE_HORIZONTAL_UM
bar_px = min(bar_px, width * 0.45)
bar_x0 = width * 0.58
bar_y = height * 0.90
bar_x1 = bar_x0 + bar_px
ax.plot([bar_x0, bar_x1], [bar_y, bar_y], color=SCALE_BAR_COLOR, linewidth=3.0, solid_capstyle="butt")
ax.text(
    (bar_x0 + bar_x1) / 2,
    bar_y - height * 0.035,
    f"{SCALE_BAR_UM:g} ?m",
    color="white",
    ha="center",
    va="bottom",
    fontsize=12,
    fontweight="bold",
)

ax.set_title(layer_name)
ax.set_xlabel("x [pixel]")
ax.set_ylabel("y [pixel]")

out = OUTPUT_DIR / f"{layer_name}_annotated_dislocation_crop.png"
plt.tight_layout()
plt.savefig(out, dpi=300, bbox_inches="tight")
plt.show()
print("saved:", out.resolve())
```

## Annotated whole-frame figure

This makes the annotated figure on the full stretched 4th rocking image, using the saved bounding box coordinates. The ellipse is placed around the saved box by default, and guide lines are specified in full-image pixel coordinates.

```python
from matplotlib.patches import Ellipse, Rectangle

# Pick a layer. If None, use the first saved layer.
LAYER_TO_PLOT = None

# Whole-frame annotation controls. Coordinates are full-image pixel coordinates.
WHOLE_FRAME_SCALE_BAR_UM = 11.5
WHOLE_FRAME_SCALE_BAR_COLOR = "red"
GUIDE_COLOR = "white"
ELLIPSE_COLOR = "lightgray"
SHOW_SAVED_BBOX = False

selected = saved[0] if LAYER_TO_PLOT is None else next(s for s in saved if s["layer_name"] == LAYER_TO_PLOT)
layer_name = selected["layer_name"]
bbox = selected["bbox_xyxy"]
x0, y0, x1, y1 = bbox

preview = next(p for p in layer_previews if p["layer_dir"].name == layer_name)
image = preview["image"].astype(np.float32)

# Display full frame in log scale.
if DISPLAY_LOG_SCALE:
    display_image = np.log1p(np.clip(image, a_min=0.0, a_max=None))
else:
    display_image = image.copy()

vmin, vmax = np.percentile(display_image, [1.0, 99.5])
display_image = np.clip(display_image, vmin, vmax)

fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(display_image, cmap=DISPLAY_CMAP, vmin=vmin, vmax=vmax)

# Default ellipse around saved dislocation box.
box_cx = 0.5 * (x0 + x1)
box_cy = 0.5 * (y0 + y1)
box_w = x1 - x0
box_h = y1 - y0
ellipse = Ellipse(
    (box_cx, box_cy),
    width=max(box_w * 1.35, 20),
    height=max(box_h * 1.35, 20),
    angle=-20,
    fill=False,
    edgecolor=ELLIPSE_COLOR,
    linewidth=3.0,
)
ax.add_patch(ellipse)

if SHOW_SAVED_BBOX:
    ax.add_patch(Rectangle((x0, y0), box_w, box_h, fill=False, edgecolor="red", linewidth=2.0))

# Guide lines through/near the dislocation. Tune slopes/lengths here if needed.
line_len = 420
ax.plot(
    [box_cx - 0.35 * line_len, box_cx + 0.35 * line_len],
    [box_cy + 0.55 * line_len, box_cy - 0.55 * line_len],
    color=GUIDE_COLOR,
    linewidth=1.8,
    alpha=0.95,
)
ax.plot(
    [box_cx - 0.55 * line_len, box_cx + 0.35 * line_len],
    [box_cy - 0.55 * line_len, box_cy + 0.55 * line_len],
    color=GUIDE_COLOR,
    linewidth=1.5,
    linestyle=":",
    alpha=0.85,
)
ax.axhline(box_cy, color=GUIDE_COLOR, linewidth=1.5, linestyle=":", alpha=0.85)

# Red scale bar on full frame. Horizontal pixel size is lab-y pixel size.
height, width = display_image.shape
bar_px = WHOLE_FRAME_SCALE_BAR_UM / PIXEL_SIZE_HORIZONTAL_UM
bar_px = min(bar_px, width * 0.25)
bar_x0 = width * 0.62
bar_y = height * 0.90
bar_x1 = bar_x0 + bar_px
ax.plot([bar_x0, bar_x1], [bar_y, bar_y], color=WHOLE_FRAME_SCALE_BAR_COLOR, linewidth=3.2, solid_capstyle="butt")
ax.text(
    (bar_x0 + bar_x1) / 2,
    bar_y - height * 0.035,
    f"{WHOLE_FRAME_SCALE_BAR_UM:g} ?m",
    color="white",
    ha="center",
    va="bottom",
    fontsize=13,
    fontweight="bold",
)

ax.set_title(f"{layer_name} - full frame")
ax.set_xlabel("x [pixel]")
ax.set_ylabel("y [pixel]")

out = OUTPUT_DIR / f"{layer_name}_annotated_full_frame.png"
plt.tight_layout()
plt.savefig(out, dpi=300, bbox_inches="tight")
plt.show()
print("saved:", out.resolve())
```
