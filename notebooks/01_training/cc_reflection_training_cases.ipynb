# Crystallographic-constrained model: reflection benchmark

This notebook benchmarks only the crystallographic-constrained model in three cases:

1. train/test on `int_fcc_002`
2. train/test on `int_fcc_111`
3. train/test on the combined `int_fcc_002 + int_fcc_111` datasets

It reports accuracy, macro metrics, training curves, and percentage confusion matrices.


```python
from dataclasses import replace
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from bv_repro.training import BenchmarkConfig, run_crystallographic_only_benchmark
```

```python
DATASET_002 = Path("/data/projects/engage_id03/int_fcc_002/")
DATASET_111 = Path("/data/projects/engage_id03/int_fcc_111/")

TRAINING_CASES = {
    "FCC 002 only": [DATASET_002],
    "FCC 111 only": [DATASET_111],
    "FCC 002 + 111": [DATASET_002, DATASET_111],
}

BASE_OUTPUT = Path("cc_reflection_only_benchmark")
BASE_OUTPUT.mkdir(parents=True, exist_ok=True)

BASE_CONFIG = BenchmarkConfig()

# Set a smaller value for a quick test, or leave None for BenchmarkConfig.epochs.
OVERRIDE_EPOCHS = None
```

```python
results = {}

for case_name, data_paths in TRAINING_CASES.items():
    case_dir = BASE_OUTPUT / case_name.lower().replace(" ", "_").replace("+", "and")
    config = replace(
        BASE_CONFIG,
        artifact_dir=str(case_dir),
        epochs=OVERRIDE_EPOCHS if OVERRIDE_EPOCHS is not None else BASE_CONFIG.epochs,
    )

    print("\n" + "=" * 80)
    print("CC-only case:", case_name)
    print("Datasets:", [str(path) for path in data_paths])
    print("=" * 80)

    results[case_name] = run_crystallographic_only_benchmark(
        data_paths=data_paths,
        config=config,
    )
```

```python
metric_rows = []
for case_name, result in results.items():
    row = result["metrics"]
    metric_rows.append({
        "Training case": case_name,
        "Accuracy (%)": row["accuracy_percent"],
        "Macro Precision (%)": row["macro_precision_percent"],
        "Macro Recall (%)": row["macro_recall_percent"],
        "Macro F1 (%)": row["macro_f1_percent"],
    })

metrics_df = pd.DataFrame(metric_rows)
display(metrics_df.style.format({
    "Accuracy (%)": "{:.2f}",
    "Macro Precision (%)": "{:.2f}",
    "Macro Recall (%)": "{:.2f}",
    "Macro F1 (%)": "{:.2f}",
}))

metrics_df.to_csv(BASE_OUTPUT / "cc_reflection_metrics.csv", index=False)
```

```python
# Combined publication-style benchmark figure: rows = reflection cases,
# columns = loss, accuracy, normalized confusion matrix.
PLOT_CASES = ["FCC 002 only", "FCC 111 only"]
case_items = [(name, results[name]) for name in PLOT_CASES if name in results]
num_cases = len(case_items)

fig, axes = plt.subplots(
    num_cases,
    3,
    figsize=(11.5, 3.2 * num_cases),
    gridspec_kw={"width_ratios": [1.0, 1.0, 1.25]},
    constrained_layout=True,
)

if num_cases == 1:
    axes = axes[np.newaxis, :]

confusion_images = []

for row_index, (case_name, result) in enumerate(case_items):
    history = result["history"]
    epochs = np.arange(1, len(history["train_loss"]) + 1)

    # Loss
    ax_loss = axes[row_index, 0]
    ax_loss.plot(
        epochs,
        history["train_loss"],
        color="blue",
        marker="s",
        markersize=3.5,
        linewidth=1.4,
        label="Train",
    )
    ax_loss.plot(
        epochs,
        history["test_loss"],
        color="red",
        marker="o",
        markersize=3.5,
        linewidth=1.2,
        linestyle="--",
        label="Valid",
    )
    ax_loss.set_xlabel("Epoch")
    ax_loss.set_ylabel("Loss")
    ax_loss.grid(True, linestyle="--", alpha=0.35)
    if row_index == 0:
        ax_loss.set_title("Loss")
        ax_loss.legend(frameon=False, fontsize=8)

    # Accuracy
    ax_accuracy = axes[row_index, 1]
    ax_accuracy.plot(
        epochs,
        np.asarray(history["train_accuracy"]) * 100.0,
        color="blue",
        marker="s",
        markersize=3.5,
        linewidth=1.4,
        label="Train",
    )
    ax_accuracy.plot(
        epochs,
        np.asarray(history["test_accuracy"]) * 100.0,
        color="red",
        marker="o",
        markersize=3.5,
        linewidth=1.2,
        linestyle="--",
        label="Valid",
    )
    ax_accuracy.set_xlabel("Epoch")
    ax_accuracy.set_ylabel("Accuracy (%)")
    ax_accuracy.set_ylim(0, 102)
    ax_accuracy.grid(True, linestyle="--", alpha=0.35)
    if row_index == 0:
        ax_accuracy.set_title("Accuracy")

    # Confusion matrix stored as percentages; convert to fractions for 0.00-1.00 labels.
    ax_cm = axes[row_index, 2]
    cm = np.asarray(result["confusion_percent"], dtype=float) / 100.0
    num_classes = cm.shape[0]
    bv_labels = [f"BV {index}" for index in range(num_classes)]

    image = ax_cm.imshow(cm, cmap="Blues", vmin=0.0, vmax=1.0, aspect="equal")
    confusion_images.append(image)
    ax_cm.set_xticks(np.arange(num_classes))
    ax_cm.set_yticks(np.arange(num_classes))
    ax_cm.set_xticklabels(bv_labels, fontsize=8)
    ax_cm.set_yticklabels(bv_labels, fontsize=8)
    ax_cm.set_xlabel("Predicted BV")
    ax_cm.set_ylabel("True BV")
    if row_index == 0:
        ax_cm.set_title("Confusion matrix")

    for i in range(num_classes):
        for j in range(num_classes):
            value = cm[i, j]
            if value >= 0.005:
                ax_cm.text(
                    j,
                    i,
                    f"{value:.2f}",
                    ha="center",
                    va="center",
                    color="white" if value >= 0.55 else "black",
                    fontsize=7.5,
                )

    # Row label identifies the reflection-training case.
    ax_loss.text(
        -0.38,
        0.5,
        case_name,
        transform=ax_loss.transAxes,
        rotation=90,
        ha="center",
        va="center",
        fontsize=10,
        fontweight="bold",
    )

# One shared colorbar for all confusion matrices.
cbar = fig.colorbar(
    confusion_images[-1],
    ax=axes[:, 2],
    fraction=0.035,
    pad=0.03,
)
cbar.set_label("Fraction of true class")
cbar.set_ticks(np.linspace(0, 1, 6))

output_path = BASE_OUTPUT / "cc_reflection_benchmark_3x3.png"
fig.savefig(output_path, dpi=300, bbox_inches="tight")
print("saved:", output_path.resolve())
plt.show()
```
